```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_base = pd.read_csv("spambase.csv") #leitura do arquivo base
```

```python
#Primeiro vamos verificar algumas inofrmações importantes do data frame, fazemos isso com algumas funções especiais do pandas
#Sendo elas .head(), .info(), .describe(), .shape and .isnull().sum()
#Comecemos pela função mais importante que nos retornará o número de valores nulos em cada coluna.
#Isso é importante pois os valores nulos interferem nossa analise.
data_base.isnull().sum()
#Vamos substituir o número de valores nulus por valores 0
data_without_null = data_base.fillna(value = 0)
```

```python
data = data_without_null.drop('Unnamed: 0', axis = 1) #limpamos a primeria coluna, pois ela é só uma repetição das keys de acesso.
```

```python
colunas = list(data.columns)
```

```python
shape = tuple(data.shape)
```

```python
#temos duas colunas com tipo objeto, vamos verificar elas e tentar tranformar para o melhor tipo possivel
data["word_freq_our"] = data["word_freq_our"].replace('???', 0)
data["word_freq_our"] = data["word_freq_our"].astype('float')
```

```python
data["word_freq_000"] = data["word_freq_000"].replace('???', 0)
data["word_freq_000"] = data["word_freq_000"].astype('float')
```

```python
data["word_freq_labs"] = data["word_freq_labs"].replace('no', 0)
data["word_freq_labs"] = data["word_freq_labs"].astype('float')
```

```python
data["word_freq_hpl"] = data["word_freq_hpl"].replace("none", 0)
data["word_freq_hpl"] = data["word_freq_hpl"].astype("float")
```
